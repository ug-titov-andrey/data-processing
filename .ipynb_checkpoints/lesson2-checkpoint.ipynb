```python
t=np.linspace(0,10,51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t,f, color='green')
font_title = {'family':'sans-serif', 'fontsize':14, 'color':'red'}
label_font={'family':'serif', 'fontsize':12}
plt.title("График f(t)", fontdict= font_title, loc='right')
plt.xlabel('Значения t', fontdict = label_font )
plt.ylabel('Значения f', fontdict = label_font )
plt.axis([0.5,9.5,-2.5,2.5])
plt.grid()
plt.show()
```

```python
x = np.linspace(-3,3,51)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2) 
ax1,ax2,ax3,ax4 = ax.flatten()
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
ax4.plot(x,y4)
font_title['fontsize']=12
ax1.set_title('График1', fontdict = font_title)
ax2.set_title('График2', fontdict = font_title)
ax3.set_title('График3', fontdict = font_title)
ax4.set_title('График4', fontdict = font_title)
ax1.set_xlim([-5,5])
fig.set_size_inches(8,6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```
