В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам: 'График y1', 'График y2' и т.д.

```python
x = np.linspace(-3,3,51)
```

Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения. Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?

```python
a = np.arange(0,12)
A = a.reshape(4,-1)
Atr= A.T
Atr
B=A.dot(Atr)
print(f'Размер матрицы B равен {B.size}')
print(f'Матрица имеет только {np.linalg.matrix_rank(B)} линейно независимые строки и ее определитиль равен нулю. \n Поэтому обратная матрица не существует' )
```

### Задание 12
Поменяйте его форму так, чтобы получилась квадратная матрица C. Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10. Вычислите определитель, ранг и обратную матрицу D_inv для D.

```python
ncols =  int(size/4)
C = K.reshape((ncols,-1))
D  = B + C*10
D_inv = np.linalg.inv(D)
D_det = np.linalg.det(D)
D_rank = np.linalg.matrix_rank(D)
print(f'Определитель матрицы D равен {D_det} \n Ранг матрицы D равен {D_rank} \n Обратная матрица - {D_inv}')
```
